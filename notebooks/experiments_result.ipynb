```python
import warnings

warnings.filterwarnings("ignore")

import sys
import os

sys.path.append(os.path.abspath("../"))

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use("bmh")
```

```python
EXPERIMENT_RESULT_PATH = "../data/runs.csv"
```

```python
result = pd.read_csv(EXPERIMENT_RESULT_PATH)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(14, 6))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

sns.lineplot(x="alpha", y="within_ATT", hue="split_outnet", data=result, ax=axs[0, 0])
sns.lineplot(x="alpha", y="outof_ATT", hue="split_outnet", data=result, ax=axs[1, 0])
axs[0, 0].set_xscale("log")
axs[1, 0].set_xscale("log")
axs[0, 0].axhline(y=1676.3426, color="black", linestyle="--")
axs[1, 0].axhline(y=1676.3426, color="black", linestyle="--")

sns.lineplot(x="alpha", y="within_RMSE", hue="split_outnet", data=result, ax=axs[0, 1])
sns.lineplot(x="alpha", y="outof_RMSE", hue="split_outnet", data=result, ax=axs[1, 1])
axs[0, 1].set_xscale("log")
axs[1, 1].set_xscale("log")

sns.lineplot(x="alpha", y="within_IPM", hue="split_outnet", data=result, ax=axs[0, 2])
sns.lineplot(x="alpha", y="outof_IPM", hue="split_outnet", data=result, ax=axs[1, 2])
axs[0, 2].set_xscale("log")
axs[1, 2].set_xscale("log")
axs[0, 2].set_yscale("log")
axs[1, 2].set_yscale("log")

plt.show()
```
